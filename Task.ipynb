### **Strategy for Trading**

**Golden Cross**:
Buy when the 50-day MA crosses above the 200-day MA.

**Death Cross**:
Sell when the 50-day MA crosses below the 200-day MA.

We take a ***Long Position*** at the Golden Cross, and a ***Short Position*** at the Death Cross

```python
df['50_MA'] = df['Close'].rolling(window=50, min_periods=1).mean()
```

```python
df['200_MA'] = df['Close'].rolling(window=200, min_periods=1).mean()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Close', color='blue')
plt.plot(df.index, df['50_MA'], label='50-day MA', color='red')
plt.plot(df.index, df['200_MA'], label='200-day MA', color='green')

plt.title('Nifty 50 Closing Prices with Moving Averages (Mar 2019 - Mar 2024)')
plt.xlabel('Date')
plt.ylabel('Price (₹)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['Golden_Cross'] = (df['50_MA'] > df['200_MA']) & (df['50_MA'].shift(1) <= df['200_MA'].shift(1))
```

```python
df['Death_Cross'] = (df['50_MA'] < df['200_MA']) & (df['50_MA'].shift(1) >= df['200_MA'].shift(1))
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='Close', color='blue')
plt.plot(df.index, df['50_MA'], label='50-day MA', color='red')
plt.plot(df.index, df['200_MA'], label='200-day MA', color='green')

plt.scatter(df[df['Death_Cross']].index, df[df['Death_Cross']]['Close'], marker='v', color='red', label='Death Cross', s=200)
plt.scatter(df[df['Golden_Cross']].index, df[df['Golden_Cross']]['Close'], marker='^', color='yellow', label='Golden Cross', s=200)

plt.title('Nifty 50 Closing Prices with Moving Averages and Crosses (Mar 2019 - Mar 2024)')
plt.xlabel('Date')
plt.ylabel('Price (₹)')
plt.legend()
plt.grid(True)
plt.show()
```

### **Backtesting**

```python
initial_capital = float(1000000)
```

```python
strategy_df = pd.DataFrame(index=df.index)
strategy_df['Position'] = 0
```

```python
# Go long (1) when Golden Cross happens and short (-1) when Death Cross happens
strategy_df['Position'] = df['Golden_Cross'].apply(lambda x: 1 if x else 0) - df['Death_Cross'].apply(lambda x: 1 if x else 0)
strategy_df['Position'] = strategy_df['Position'].ffill()
```

```python
# Calculate daily market returns and strategy returns
strategy_df['Market_Returns'] = df['Close'].pct_change()
strategy_df['Strategy_Returns'] = strategy_df['Market_Returns'] * strategy_df['Position'].shift(1)
```

```python
# Calculate the cumulative returns
strategy_df['Cumulative_Market_Returns'] = (1 + strategy_df['Market_Returns']).cumprod()
strategy_df['Cumulative_Strategy_Returns'] = (1 + strategy_df['Strategy_Returns']).cumprod()
```

```python
# Apply the initial capital to the cumulative strategy returns
strategy_df['Strategy_Equity'] = initial_capital * strategy_df['Cumulative_Strategy_Returns']
```

```python
# The final equity value of the strategy
final_equity = strategy_df['Strategy_Equity'].iloc[-1]
```

```python
final_equity
```

### **Conclusion**

This means, that the strategy is **not working**. 

If carefully analysed, there was a big loss during the **Covid drop**.

This was because Moving Average is a **Lagging indicator**, and can take time to realise the market trends.
